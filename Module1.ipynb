```python


# Загрузка модели для вопросно-ответного поиска
qa_pipeline = pipeline("question-answering", model="bert-large-uncased-whole-word-masking-finetuned-squad")

# Текст модуля (подставьте ваш текст)
module_text = """
Приветствуем вас на дисциплине, посвящённой изучению анализа естественного языка. Для начала разберёмся, что из себя представляет это направление.
Текстовые данные — неотъемлемая часть нашей жизни. От социальных медиа до новостных порталов, от общения с виртуальными помощниками до анализа обратной связи клиентов — текстовая информация окружает нас повсюду.
Но каким образом мы можем извлекать ценные знания, понимать настроение и потребности людей, а также автоматически обрабатывать и интерпретировать огромные объёмы текста? В этом нам и помогает анализ естественного языка.
Что такое анализ естественного языка
На заре появления первых ЭВМ зародилась идея научить компьютер понимать естественный язык. Под естественным языком мы будем понимать любой язык, используемый для общения людей, который не был создан целенаправленно.
Многие свойства и особенности естественных языков упрощают их использование с точки зрения человека: зависимость значений слов от контекста, даже не явно выраженного, пропуски определённых слов и т. д. Они делают задачу автоматизации обработки естественного языка крайне сложной.
Направление на стыке искусственного интеллекта и лингвистики, изучающее методы автоматизации анализа текстов на естественном языке, называется Natural Language Processing (NLP) или анализом естественного языка.
Основная цель анализа естественного языка — сделать текстовые данные доступными для компьютерной обработки, интерпретации и использования.
NLP-специалисты обучают компьютеры понимать и работать с языком так, как мы, люди. Компьютеры могут распознавать текст, определять, о чём речь, и даже генерировать свои собственные тексты. Это помогает им делать разные полезные вещи, например, автоматически переводить тексты на другие языки, находить информацию в больших текстовых данных и многое другое.
Что вас ждёт в этом модуле
В этом модуле мы рассмотрим основные задачи, которые решаются в области NLP, и то, как они могут применяться на практике:
* Задача классификации текстов в целом Определение тематики текста, определение спама, определение грамматически корректных предложений, определение связанности двух предложений и т. д.
* Задача классификации отдельных элементов текста Выделение именованных сущностей в тексте, определение частей речи слов текста и т. д.
* Задача поиска ответа на вопрос Задача поиска ответа на сформулированный пользователем вопрос в заданном контексте.
* Задача генерации текста Машинный перевод, реферирование текста, заполнение пропусков в тексте, чат-боты и т. д.
* В этом модуле нас будет интересовать постановка задачи и итоговый результат без погружения в подробности реализации решения. Поэтому примеры решения задач мы приведём с использованием высокоуровневого класса pipeline из библиотеки huggingface transformers, наилучшим образом подходящего для демонстрации возможностей NLP-моделей без погружения в детали.В следующем юните мы начнём знакомить вас с классификацией текстов. Мы рассмотрим конкретную задачу классификации текстов, а в конце юнита вас ждёт реальная задача: запустить модель для произвольных текстов на английском языке и проанализировать возвращаемые результаты.
В этом юните мы расскажем о распространнённой задаче в анализе естественного языка — классификации текстов. Она представляет собой процесс присвоения той или иной категории отдельно взятой части текста.
Рассмотрим несколько примеров:
* тематическая классификация текстов по рубрикам, используемая новостными агрегаторами;
* sentiment analysis — классификация текста или его фрагментов на позитивные или негативные высказывания, например, для обработки отзывов о товарах и услугах;
* спам-фильтрация, активно применяемая для анализа e-mail и т. д.

￼

Одна из основных задач в NLP
Давайте в общих чертах вспомним pipeline решение задачи классификации с помощью алгоритмов машинного обучения на табличных данных.
Для применения какого-либо метода машинного обучения с учителем (gradient boosting, support vector machines и т. д.) нужно сформировать признаковое пространство, в котором представлены объекты обучающей и тестовой выборок.
В случае табличных данных мы в явном виде имеем матрицу объект-признак. Однако необходимо привести в числовой вид переменные, значения которых в данных представлено не числами. Есть несколько способов кодирования такого рода переменных: one-hot encoding, label encoding, embeddings и т. д.
Теперь подробнее рассмотрим задачу классификации текстов. Глобально она не сильно отличается от классификации табличных данных.
Нам также необходимо сформировать признаковое пространство, а затем обучить некоторый алгоритм классификации. Здесь наша основная задача — понять, как можно представить текст в виде вектора, чтобы можно было обучить алгоритм машинного обучения. При этом важно не просто представить текст в виде вектора, а сделать это так, чтобы учитывались особенности естественных языков, в частности, многозначность слов, наличие подразумеваемых слов и т. д.
Именно эта задача является одной из основных в NLP, и прорыв в качестве решения NLP-задач обусловлен именно новыми подходами к отображению текстов в векторное пространство. Однако задача тематической генерации достаточно успешно решалась с использованием довольно простых статистических методов моделирования текста, которые будут рассмотрены в следующих модулях курса.
Пример тематической классификации текстов
Приведём пример работы модели тематической классификации текстов.
Для того чтобы воспользоваться pipeline, нам необходимо указать два аргумента — решаемую задачу и модель, которую мы будем использовать для её решения. Со списком поддерживаемых задач этого инструмента можно ознакомиться в документации. В нашем случае нам необходимо использовать значение аргумента task="text-classification".
Как выбрать подходящую модель? На сайте Hugging Face выложено большое количество различных моделей для различных задач. С некоторыми мы познакомим вас в рамках нашего курса, а сейчас воспользуемся уже обученными моделями под нашу задачу. Для этого выбираем в фильтре task значение "Text classification", а в фильтре language выбираем английский.
￼
Источник: huggingface.co/models

Затем, учитывая, что мы хотим использовать модель для тематической классификации, нам надо выбрать модели, обученные на соответствующих датасетах.
Выберем модели, обученные на популярном датасете для тематической классификации, также с помощью соответствующего фильтра, — Ag-News, в котором представлены тексты четырёх классов:
* World,
* Sports,
* Business,
* Sci/Tech.
Например, выберем модель "nateraw/bert-base-uncased-ag-news". Но можно выбрать и любую другую из подходящих нам критериев.
Теперь всё готово для запуска! Классифицируем текст "NLP is a field of linguistics and machine learning focused on understanding everything related to human language."
cls_pl = pipeline("text-classification", "nateraw/bert-base-uncased-ag-news")
cls_pl('NLP is a field of linguistics and machine learning focused on understanding everything related to human language.')


[{'label': 'Sci/Tech', 'score': 0.9625065326690674}]


На выходе мы получили класс 'Sci/Tech' для данного текста. На параметр 'score' можно пока не обращать внимания, но неформально его можно интерпретировать, как вероятность принадлежности объекта к тому или иному классу по мнению модели.
Тренировочное задание
Попробуйте запустить эту модель для произвольных текстов на английском языке и проанализировать возвращаемые результаты. Например, можно взять следующие тексты:
* "Venezuela Prepares for Chavez Recall Vote Supporters and rivals warn of possible fraud; government says Chavez's defeat could produce turmoil in world oil market".
* "Retailers Vie for Back-To-School Buyers (Reuters) Reuters - Apparel retailers are hoping their\back-to-school fashions will make the grade among\style-conscious teens and young adults this fall, but it could\be a tough sell, with students and parents keeping a tighter\hold on their wallets".
* "Another Major Non-Factor Another major, another disappointment for Tiger Woods, the No. 1 ranked player in the world who has not won a major championship since his triumph at the 2002 U.S. Open".
Так как в этом случае вам не нужно обучать модель и вы запускаете её на очень небольшом множестве примеров, вы можете производить запуск и без использования GPU.

Естественным развитием задач классификации всего текста или его фрагментов является классификация отдельных слов в предложении.
Классификация отдельных слов лежит в основе многих приложений, которые мы используем ежедневно, начиная от поисковых систем и заканчивая системами машинного перевода и анализа социальных медиа.
Наиболее распространённая задача такого вида — задача выделения именованных сущностей (NER, Named Entity Recognition). В этом случае необходимо для каждого слова в тексте проставить класс принадлежности его к тому или иному типу сущности, например, организация, персона, локация, либо пустой класс, показывающий, что слово не относится ни к одному из рассматриваемых классов.
С помощью NER можно разрабатывать системы автоматического заполнения форм документов, составления отчётов о компаниях, нахождения разных упоминаний одних и тех же сущностей, удаления персональной информации из текстов. Кроме классификации слов текста на именованные сущности, можно производить классификацию по каким-то другим критериям, например, по частям речи, хотя такое практическое использование этого рода задач гораздо уже.
￼

Обратимся к уже знакомому классу pipeline для решения задачи NER.
В данном случае нам понадобятся модели, обученные на корпусе CoNLL2003, — одном из базовых бенчмарков для задачи NER. В этом корпусе присутствуют сущности следующих типов:
* PER (person),
* LOC (location),
* ORG (organization),
* MISC (miscellaneous).
Для данной задачи в аргументе task в pipeline необходимо передать "token-classification", либо "ner". В качестве модели выберем "dslim/bert-base-NER", но вы можете воспользоваться и другой, подходящей по соответствующим фильтрам. Также нам понадобится аргумент aggregation_strategy="first" для обеспечения более читаемого вывода из-за особенности токенизации слов в данной задаче, которую мы рассмотрим позднее.
Попробуем выделить именованные сущности из текста "Sylvain Gugger is a Research Engineer at Hugging Face and one of the core maintainers of the Hugginface Transformers library".

Одной из самых первых задач обработки естественного языка исторически была задача машинного перевода. Достаточно долгое время автоматизированные переводчики не могли добиться удовлетворительного качества. Дословный перевод текста, без учёта контекста и особенностей рассматриваемых языков, может служить разве что отправной точкой для дальнейшей ручной доработки перевода.
Даже современные решения данной задачи могут по-прежнему не всегда показывать хорошие результаты, но задача автоматизации перевода остаётся очень востребованной на практике и используется в явном виде — для разработки систем перевода между естественными языками.
￼

Попробуем решить задачу машинного перевода с помощью уже знакомого нам инструмента от hugginface. Переведём текст с английского на русский.
В качестве аргумента task зададим "translation_en_to_ru".
В качестве модели используем "facebook/wmt19-en-ru", которую вы можете заменить на любую другую, подходящую по соответствующим фильтрам, и подадим на вход текст «Matthew Carrigan is a Machine Learning Engineer at Hugging Face. He lives in Dublin, Ireland and previously worked as an ML engineer at Parse.ly and before that as a post-doctoral researcher at Trinity College Dublin».
В итоге код будет выглядеть следующим образом:
txt = "Matthew Carrigan is a Machine Learning Engineer at Hugging Face. He lives in Dublin, Ireland and previously worked as an ML engineer at Parse.ly and before that as a post-doctoral researcher at Trinity College Dublin."

en_ru_translator = pipeline("translation_ru_to_en", "facebook/wmt19-en-ru")
en_ru_translator(txt)

Мы получили текст: «Мэтью Кэрриган — инженер-механик в компании Hugging Face. Он живет в Дублине, Ирландия, а до этого работал инженером-программистом в Parse.ly, а до этого — постдоком-исследователем в дублинском колледже Trinity College».
Несмотря на то что Мэтью был бы очень удивлён, узнав, что он работает инженером-механиком, перевод получился вполне достойным — соблюдена грамматическая структура, а имена собственные, которые не надо переводить, оставлены на языке оригинала.
Попробуем перевести полученный результат обратно на английский. Для этого нам понадобится уже другая модель и другой аргумент task — "facebook/wmt19-ru-en" и "translation_ru_to_en" соответственно.
ru_en_translator = pipeline("translation_ru_to_en", "facebook/wmt19-ru-en")

ru_en_translator("Мэтью Кэрриган - инженер-механик в компании Hugging Face. Он живет в Дублине, Ирландия, а до этого работал инженером-программистом в Parse.ly, а до этого - постдоком-исследователем в дублинском колледже Trinity College.")

В итоге мы получили текст: «Matthew Kerrigan is a mechanical engineer at Hugging Face. He lives in Dublin, Ireland, and previously worked as a software engineer at Parse.ly and as a postdoctoral researcher at Trinity College in Dublin». Вполне неплохой результат.
Техника, которой мы воспользовались, часто используется в NLP-соревнованиях, на хакатонах для аугментации данных: с помощью моделей автоматического перевода тексты обучающей выборки переводят на некоторый язык, а затем обратно на язык оригинала. Тем самым в большинстве случаев получается очень близкий к оригиналу текст, но содержащий незначительные изменения. Это может оказаться очень полезным, например, для расширения обучающей выборки.

"""

# Вопросы
questions = [
    "Какая задача может быть развитием задачи классификации текстов?",
    "Что послужило основным толчком для существенного улучшения качества решения NLP-задач в последние несколько лет?",
    "Что служит входными данными для задачи поиска ответа на вопрос в заданном контексте?",
    "Какие NLP-задачи успешно решались с использованием классических статистических методов?",
]

# Поиск ответов на вопросы в тексте модуля
answers = []
for question in questions:
    answer = qa_pipeline(question=question, context=module_text)
    answers.append(answer)

# Вывод ответов
for i, answer in enumerate(answers):
    print(f"Вопрос {i + 1}: {questions[i]}")
    print(f"Ответ: {answer['answer']}\n")
```
